```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
print("Missing values in each column:")

print(df.isna().sum())

print("Number of duplicate rows:", df.duplicated().sum())
```

```python
import pandas as pd

# 1. Year range
min_year = df['year'].min()
max_year = df['year'].max()
print(f"Year range: {min_year} to {max_year}")

# 2. Number of countries
# nunique() counts the number of distinct/unique values in the column
num_countries = df['country'].nunique()
print(f"Number of unique countries: {num_countries}")

# 3. Number of continents
num_continents = df['continent'].nunique()
print(f"Number of unique continents: {num_continents}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Get the data
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# 2. Calculate the average and draw the chart all in one step
df.groupby('continent')['lifeExp'].mean().plot(kind='bar')

# 3. Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 2. Calculate the averages and draw the line graph in one step
df.groupby(['year', 'continent'])['lifeExp'].mean().unstack().plot()

# 3. Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 2. Draw a simple scatter plot
df.plot(kind='scatter', x='gdpPercap', y='lifeExp')

# 3. Show the chart
plt.show()
```
