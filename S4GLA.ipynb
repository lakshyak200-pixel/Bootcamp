```python
#calculate the average score accroess the other two subjects

#filtering

#top students average more than or equal to 85

df["average"] = df[["math", "science", "english"]].mean(axis=1)

print(df[["name", "math", "science", "english", "average"]])

top_students =df [df["average"] >= 85]

print(top_students[["name", "average"]])

```

```python

# 1. Filter for low attendance (less than 75%)
low_attendance = df[df["attendance_percent"] < 75]

print("Students with Low Attendance")
print(low_attendance[["name", "attendance_percent"]])


# 2. Filter for students from Mathura and show specific columns
mathura_students = df[df["city"] == "Mathura"]

print("Students from Mathura")
print(mathura_students[["name", "city", "average"]])
```

```python
import pandas as pd
import numpy as np

# 2. Create an 'average' column using math, science, and english
df["average"] = df[["math", "science", "english"]].mean(axis=1)

# 3. Create a 'result' column: Pass if average >= 40, otherwise Fail
df["result"] = np.where(df["average"] >= 40, "Pass", "Fail")

# 4. Print class average
class_average = df["average"].mean()
print(f"Class Average{class_average:.2f}\n")

# 5. Print highest average and lowest average
highest_avg = df["average"].max()
lowest_avg = df["average"].min()
print("Highest and Lowest Averages")
print(f"Highest Average: {highest_avg:.2f}")
print(f"Lowest Average: {lowest_avg:.2f}\n")

# 6. Print top student or students
top_students = df[df["average"] == highest_avg]
print("Top Student(s)")
print(top_students[["name", "average"]])
print("\n")

# 7. Print students with attendance below 75
low_attendance = df[df["attendance_percent"] < 75]
print("Students with Attendance Below 75%")
print(low_attendance[["name", "attendance_percent"]])
print("\n")

# 8. Print students who submitted fewer than 6 assignments
few_assignments = df[df["assignments_submitted"] < 6]
print("Students with Fewer than 6 Assignments")
print(few_assignments[["name", "assignments_submitted"]])
print("\n")

# 9. Save final output as session4_student_report.csv
output_filename = "session4_student_report.csv"
df.to_csv(output_filename, index=False)
print(f"Success! Report saved to {output_filename}")
```
